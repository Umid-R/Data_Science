```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
df=pd.read_csv('without_null_cleaned_data.csv')
```

```python
df.head()
```

```python
# Countplot

sns.set_theme(style="darkgrid")
sns.countplot(df,x='Status')
```

```python
# Histplot 
sns.histplot(df,x='Monthly Salary (INR)',kde=True)
```

```python
# Countplot
sns.countplot(df,x='Education', hue='AI Risk')
```

```python
# Boxplot
age_order = ['18-24', '25-34', '35-44', '45-54', '55+']
sns.boxplot(df,x='Age Group',y='Monthly Salary (INR)',order=age_order)
```

```python
# Barplot
sns.barplot(df,x='Location',y='Monthly Salary (INR)',estimator=np.mean)
```

```python
# Heatmap
sns.heatmap(df.select_dtypes('number').corr())
```

```python
# Scatterplot
sns.relplot(df, x='Years of Experience',y='Monthly Salary (INR)',hue='Industry',size='Monthly Salary (INR)',col='Location',kind='scatter',col_wrap=2)
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Data Cleaning Status

df['Status'] = df['Status'].str.strip().str.lower()
df['Status']=df['Status'].replace({
    'employed': 'Employed',
    'unemployed': 'Unemployed'
}) 



    
```

```python
# Data cleaning Age Group
df['Age Group'].str.strip()
df['Age Group']=df['Age Group'].str.replace(' ','')
df['Age Group']=df['Age Group'].str.replace('_','-')
df['Age Group']=df['Age Group'].str.replace('to','-')
```

```python
# Data Cleaning Education
df['Education']=df['Education'].str.lower().str.strip()
df['Education']=df['Education'].replace({
    'high school':'High school',
    'master':"Master's",
    "masters":"Master's",
    'phd':'PhD',
    'ph.d':'PhD',
    'bachelors':"Bachelor's",
    "bachelor's":"Bachelor's",
    'diploma':'Diploma'
})
df['Education'].unique()
```

```python
# Data Cleaning Industry

df['Industry']=df['Industry'].str.strip()
df['Industry']=df['Industry'].replace({
    'Tech':'Technology',
    'Fintech':'Finance',
    'Health':'Healthcare'
})
df['Industry'].unique()
```

```python
# Data Cleaning Location
df['Location']=df['Location'].str.strip().str.lower()
df['Location']=df['Location'].replace({
    'urban':'Urban',
    'rural':'Rural',
    'delhi':"Delhi",
    'hyderabad':'Hyderabad',
    'bangalore':'Bangalore',
    'mumbai':'Mumbai',
    'suburban':'Suburban'
})
df['Location'].unique()

```

```python
# Data Cleaning AI Risk
df['AI Risk']=df['AI Risk'].str.strip().str.lower().str.capitalize()
df['AI Risk']=df['AI Risk'].replace({
    'Moderate':'Medium'
})
df['AI Risk'].unique()
```

```python
# Data Cleaning Years of Experience
df['Years of Experience']
```

```python
# Data Cleaning Monthly Salary 
df['Monthly Salary (INR)']
```

```python
# Data Cleaning Date Recorded
df['Date Recorded']=pd.to_datetime(df['Date Recorded'], format='%m/%d/%Y')
df['Date Recorded']
```

```python
# Categorical columns - fill with mode
for col in ['Status', 'Age Group', 'Education', 'Industry', 'Location', 'AI Risk']:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Years of Experience - grouped median by Age Group
df['Years of Experience'] = df.groupby('Age Group')['Years of Experience']\
                              .transform(lambda x: x.fillna(x.median()))

# Monthly Salary - grouped median by Industry and Status
df['Monthly Salary (INR)'] = df.groupby(['Industry', 'Status'])['Monthly Salary (INR)']\
                               .transform(lambda x: x.fillna(x.median()))

```

```python
df.to_csv("cleaned_data.csv",index=False)
```

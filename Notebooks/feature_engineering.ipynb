```python
import pandas as pd
```

```python
df=pd.read_csv('Datasets/cleaned_data.csv')
```

```python
df.head(5)
```

```python
# Featuring "Status"
df['Status'] = df['Status'].map({
    'Employed': 1,
    'Unemployed': 0
})
```

```python
# Featuring "Age Group"
age_map = {
    '18-24': 1,
    '25-34': 2,
    '35-44': 3,
    '45-54':4,
    '55+': 5
}

df['Age Group'] = df['Age Group'].map(age_map)
```

```python
# Featuring "Education"
education_map={
    'High school':1,
    'Diploma':2,
    "Bachelor's":3,
    "Master's":4,
    'PhD':5
}

df['Education']=df['Education'].map(education_map)
```

```python
# Featuring "Industry" and "Location"
df = pd.get_dummies(df, columns=['Industry', 'Location'], dtype=int)
```

```python
# Featuring "AI Risk"

ai_risk_map = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}

df['AI Risk'] = df['AI Risk'].map(ai_risk_map)
```

```python
# Featuring "Date Recorded"
df['Date Recorded']=pd.to_datetime(df['Date Recorded'])
df['Year']=df['Date Recorded'].dt.year
df['Month']=df['Date Recorded'].dt.month
df['Day']=df['Date Recorded'].dt.day
df.drop('Date Recorded', axis=1, inplace=True)
```
